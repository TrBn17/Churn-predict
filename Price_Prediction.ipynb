Check and clean data

```python
#Drop unneccesary col
df.drop(['Title', 'Nội Dung','Latitude','Longtitude'], axis=1, inplace=True)
```

```python
#Convert to exactly format
import re
def convert_price_to_numeric(price_str):
    if pd.isna(price_str):
        return None
    numeric_part = re.sub(r'[^\d.]+', '', price_str)
    try:
        return float(numeric_part)
    except ValueError:
        return None

df['Giá/m2'] = df['Giá/m2'].apply(convert_price_to_numeric)
df['Giá'] = df['Giá'].apply(convert_price_to_numeric)
df.head()
```

```python
#Convert to exactly format
def convert_area_to_numeric(area_str):
    if pd.isna(area_str):
        return None
    numeric_part = re.sub(r'[^\d.]+', '', area_str)
    try:
        return float(numeric_part)
    except ValueError:
        return None

df['Diện tích'] = df['Diện tích'].apply(convert_area_to_numeric)
```

```python
df
```

```python
df['Thời Gian'] = pd.to_datetime(df['Thời Gian'])
```

```python
#CHECK
df['Giá/m2'] = pd.to_numeric(df['Giá/m2'], errors='coerce')
df['Giá'] = pd.to_numeric(df['Giá'], errors='coerce')

column_types_after_conversion = df.dtypes
column_types_after_conversion
```

```python
#Fill N/A by mean
mean_price_per_m2 = df['Giá/m2'].mean()
print(mean_price_per_m2)
df['Giá/m2'].fillna(mean_price_per_m2, inplace=True)
```

```python
df['Giá'] = df['Giá/m2'] * df['Diện tích']
```
